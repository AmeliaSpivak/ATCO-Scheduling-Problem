```python
 def g(frac,dJD,TotalNumWorkers):
        
        i=0
        b=0
        d=0
        count=0
        while count< frac*TotalNumWorkers and i in range(0,34):
            j= len(dJD)-1-i
            newcount= dJD.iloc[j,0]
            count= newcount + count
            if count> frac*TotalNumWorkers:
                    d= count-frac*TotalNumWorkers
                    b= dJD.iloc[j,0]-d

            i = i + 1
    
        return j,b,d

      


def eammn_mmpnn(WorkerDistrictID):
    
    
    Remainder = len(WorkerDistrictID) % 5
    TotalNumWorkers = len(WorkerDistrictID) - Remainder
    
    np.random.seed(10)
    neighborhood_List = np.random.choice(WorkerDistrictID, len(WorkerDistrictID), replace=False)
    
    
    Neighborhoodlist = neighborhood_List[0:TotalNumWorkers]
    #np.array(Neighborhoodlist)

    
    

    NeighborhoodSeries = pd.Series(Neighborhoodlist)
    dJD=EAMMN_MMPNN.copy()
    dJD.insert(0,'NumWorkers', NeighborhoodSeries.value_counts())
    dJD=dJD.fillna(0)
    
    
    #We will first fill up the three shifts of EAMMN_MMPNN having the lowest total commute time, EM, NN, MN, with workers from
    #districts that take the greatest values during these shifts, i.e. districts with the greatest values of DIFFMAX. That is,
    #we will count out starting from the bottom of EAMMN_MMPNN, 3/5*TotalNumWorkers who we will assign to EM, NN, MN, in the 
    # manner indicated below.
  
    
    j,b,d = g(3/5,dJD,TotalNumWorkers)
    
    
    
    #In counting 3/5*TotalNumWorkers workers starting from the bottom of EAMMN_MMPNN and working upwards, sometimes the number 
    #3/5*TotalNumWorkers may include some of the workers in the last district counted but not all of the workers in that district.
    #So that some of the workers in the last district considered are in the 3/5*TotalNumWorkers and therefore to be assigned
    #to EM, NN or MN, and the remainder of workers in that district will be assigned to the other two shifts.
    # 'b' denotes the number of workers in that last district that are counted in 3/5*TotalNumWorkers, and 'd' denotes the 
    # number of workers remaining in that last district that are not counted in 3/5*TotalNumWorkers. 'b+d' is the sum of the 
    # workers in that last district reached in counting 3/5*TotalNumWorkers.
    # The data frame EAMMN_MMPNN is broken up into two data frames: Low3 contains the 3/5*TotalNumWorkers workers who will be 
    # assigned shifts EM, NN, and MN; uppdJD wil contain the 2/5*TotalNumWorkers left which will include the d workers descrribed
    # above.
    
    Low3= dJD[j:35].copy()
    
    
    if b>0:
        Low3.iloc[0,0]=b
    
    uppdJD= dJD[0:j+1]
    uppdJD= uppdJD.copy()
   
    
    if d>0:
        uppdJD.iloc[j,0]= d
    else:
        uppdJD.iloc[j,0]=0

    
    #Filling up shifts EM and NN, which have the lowest commute values first, from the 3/5*TotalNumWorkers workers.
    # We first assign each district to the shift, EM or NN, for which it has the lower commute value. This often results in too 
    # many, i.e. more than 1/5*TotalNumWorkers being assigned to one o the two shifts and then some of those extra workers need
    # to be redistributed either to the other shift if it has been filled with fewer than 1/5*TotalNumWorkers or to MN.

    
    EM={}
    NN={}
    for i in range(0,len(Low3)):
        a=Low3.iloc[i,1]
        b=Low3.iloc[i,0]
        if Low3.iloc[i,2]<Low3.iloc[i,4]:
            EM[a]=b
            
        else:
            NN[a]=b
        i = i + 1
        
    DiffEMandMN= Low3['MtoN']-Low3['EtoM']
   
  

    MN={}
    EMnearMN=DiffEMandMN[list(EM.keys())].sort_values()

   
    y=sum(EM.values())-1/5*TotalNumWorkers
    for i in range(0,len(EMnearMN)):
        x=EMnearMN.index[i]
        z=EM[x]
        w=y-z
        if w<0:

            EM[x]=z-y
            MN[x]=y
            break
        else:
            y=y-z
            MN[x]=z
            del EM[x]
    DiffNNandMN= Low3['MtoN']-Low3['NtoN']
   
  

    ADDtoMN={}
    NNnearMN=DiffNNandMN[list(NN.keys())].sort_values()

    
    y=sum(NN.values())-1/5*TotalNumWorkers
    for i in range(0,len(NNnearMN)):
        x=NNnearMN.index[i]
        z=NN[x]
        w=y-z
        if w<0:

            NN[x]=z-y
            ADDtoMN[x]=y
            break
        else:
            y=y-z
            ADDtoMN[x]=z
            del NN[x]
   
    MN.update(ADDtoMN)
    
    neighborhood_List[TotalNumWorkers:]
    
    for i in neighborhood_List[TotalNumWorkers:]:
        if i in NN.keys():
            NN[i] = NN[i] + 1
        else:
            NN[i] = 1
    
    
    
    

    r= 1/5*TotalNumWorkers-sum(MN.values())
   
    for i in range(1,int(r)):
        
        b=dJD.iloc[len(NNnearMN)-i,0]
        a=dJD.iloc[len(NNnearMN)-i,1]
        if b>=r:
            MN[a]=r

            break
        else:
            MN[a]=b
            r= r-b

    #Having filled up EM, NN, and MN with the workers of Low3 we now turn to the uppdJD to fill out slots AM and MP. 
    
    AM={}
    MP={}
    for i in range(0,j+1):
        a=uppdJD.iloc[i,1]
        b=uppdJD.iloc[i,0]
        if uppdJD.iloc[i,3]< uppdJD.iloc[i,4]:
            AM[a]=b
        else:
            MP[a]=b

    DiffAMandMP= uppdJD['AtoM']-uppdJD['MtoP']
    AMnearMP= DiffAMandMP[list(MP.keys())].sort_values()

    DiffMPandAM= uppdJD['MtoP']- uppdJD['AtoM']
    MPnearAM=DiffMPandAM[list(AM.keys())].sort_values()

    s= sum(MP.values())-sum(AM.values())
    i=0
    
    while s!=0:
        if s>0:
            t= sum(MP.values())-1/5*TotalNumWorkers
            z= MP[AMnearMP.index[i]]
            if z>t:
                AM[AMnearMP.index[i]]=t
                MP[AMnearMP.index[i]]=z-t
                s= s-2*t
            else:
                AM[AMnearMP.index[i]]=z
                del MP[AMnearMP.index[i]]
                s= s-2*z 
        elif s<0:
            t= sum(AM.values())-1/5*TotalNumWorkers
            z= AM[MPnearAM.index[i]]
            if z>t:
                MP[MPnearAM.index[i]]=t
                AM[MPnearAM.index[i]]=z-t
                s= s+ 2*t
            else:
                MP[MPnearAM.index[i]]=z
                del AM[MPnearAM.index[i]]
                s= s +2*z
        i = i + 1

    # Having placed all the workers into the 5 slots we now compute the Total Commute Time for the workforce.
    # First we compute the total commute time for the workers in shift AM.

    TotCommAM=[]
    for i in AM.keys():
        w=dJD.loc[i,"AtoM"]
        y=AM[i]
        m= y*w
        TotCommAM.append(m)


    SumAM=sum(TotCommAM)

    # We compute the total commte time for the workers in shift MP, then MN, then EM and then NN.

    TotCommMP=[]

    for i in MP.keys():
        w=dJD.loc[i,"MtoP"]
        y=MP[i]
        m=y*w
        TotCommMP.append(m)


    SumMP=sum(TotCommMP)

    TotCommMN=[]
    for i in MN.keys():
        w=dJD.loc[i,"MtoN"]
        y=MN[i]
        m= y*w
        TotCommMN.append(m)


    SumMN=sum(TotCommMN)


    TotCommEM=[]
    for i in EM.keys():
        w=dJD.loc[i,"EtoM"]
        y=EM[i]
        m= y*w
        TotCommEM.append(m)


    SumEM=sum(TotCommEM)


    TotCommNN=[]
    for i in NN.keys():
        w=dJD.loc[i,"NtoN"]
        y=NN[i]
        m= y*w
        TotCommNN.append(m)


    SumNN=sum(TotCommNN)

    SumEAMMN_MMPNN = SumEM + SumAM + SumMP + SumMN + SumNN
    
    
    
    return SumEAMMN_MMPNN, EM, AM, MP, 0, 0, MN, NN, 0,0,0 


def eammn_ammnn(WorkerDistrictID):
    
    
    Remainder = len(WorkerDistrictID) % 5
    TotalNumWorkers = len(WorkerDistrictID) - Remainder
    
    np.random.seed(10)
    neighborhood_List = np.random.choice(WorkerDistrictID, len(WorkerDistrictID), replace=False)
    
    
    Neighborhoodlist = neighborhood_List[0:TotalNumWorkers]
    

    
    

    NeighborhoodSeries = pd.Series(Neighborhoodlist)
    dJD=EAMMN_AMMNN.copy()
    dJD.insert(0,'NumWorkers', NeighborhoodSeries.value_counts())
    dJD=dJD.fillna(0)
    
    
    #Comments
    
    j,b,d = g(3/5,dJD,TotalNumWorkers)
    
    
    
    
    #In comments explain what b and d are.
    
    Low3= dJD[j:35].copy()
    
    
    if b>0:
        Low3.iloc[0,0]=b
    
    uppdJD= dJD[0:j+1]
    uppdJD= uppdJD.copy()
   
    
    if d>0:
        uppdJD.iloc[j,0]= d
    else:
        uppdJD.iloc[j,0]=0

    
    EA={}
    NN={}
    for i in range(0,len(Low3)):
        a=Low3.iloc[i,1]
        b=Low3.iloc[i,0]
        if Low3.iloc[i,2]<Low3.iloc[i,4]:
            EA[a]=b
            
        else:
            NN[a]=b
        i = i + 1
        
    DiffEAandMN= Low3['MtoN']-Low3['EtoA']
    
   
  

    MN={}
    EAnearMN=DiffEAandMN[list(EA.keys())].sort_values()
    

   
    y=sum(EA.values())-1/5*TotalNumWorkers
    for i in range(0,len(EAnearMN)):
        x=EAnearMN.index[i]
        z=EA[x]
        w=y-z
        if w<0:

            EA[x]=z-y
            MN[x]=y
            break
        else:
            y=y-z
            MN[x]=z
            del EA[x]
    DiffNNandMN= Low3['MtoN']-Low3['NtoN']
   
  

    ADDtoMN={}
    NNnearMN=DiffNNandMN[list(NN.keys())].sort_values()

    
    y=sum(NN.values())-1/5*TotalNumWorkers
    for i in range(0,len(NNnearMN)):
        x=NNnearMN.index[i]
        z=NN[x]
        w=y-z
        if w<0:

            NN[x]=z-y
            ADDtoMN[x]=y
            break
        else:
            y=y-z
            ADDtoMN[x]=z
            del NN[x]
   
    MN.update(ADDtoMN)
    
    neighborhood_List[TotalNumWorkers:]
    
    for i in neighborhood_List[TotalNumWorkers:]:
        if i in NN.keys():
            NN[i] = NN[i] + 1
        else:
            NN[i] = 1
    
    
    
    

    r= 1/5*TotalNumWorkers-sum(MN.values())
   
    for i in range(1,int(r)):
        
        b=dJD.iloc[len(NNnearMN)-i,0]
        a=dJD.iloc[len(NNnearMN)-i,1]
        if b>=r:
            MN[a]=r

            break
        else:
            MN[a]=b
            r= r-b

    # Having filled up EM, NN, and MN with workers we now turn to the uppdJD to fill out slots AM and MP.      
    AM={}
    MM={}
    for i in range(0,j+1):
        a=uppdJD.iloc[i,1]
        b=uppdJD.iloc[i,0]
        if uppdJD.iloc[i,3]< uppdJD.iloc[i,4]:
            AM[a]=b
        else:
            MM[a]=b

    DiffAMandMM= uppdJD['AtoM']-uppdJD['MtoM']
    AMnearMM= DiffAMandMM[list(MM.keys())].sort_values()

    DiffMMandAM= uppdJD['MtoM']- uppdJD['AtoM']
    MMnearAM=DiffMMandAM[list(AM.keys())].sort_values()

    s= sum(MM.values())-sum(AM.values())
    i=0
    
    while s!=0:
        if s>0:
            t= sum(MM.values())-1/5*TotalNumWorkers
            z= MM[AMnearMM.index[i]]
            if z>t:
                AM[AMnearMM.index[i]]=t
                MM[AMnearMM.index[i]]=z-t
                s= s-2*t
            else:
                AM[AMnearMM.index[i]]=z
                del MM[AMnearMM.index[i]]
                s= s-2*z 
        elif s<0:
            t= sum(AM.values())-1/5*TotalNumWorkers
            z= AM[MMnearAM.index[i]]
            if z>t:
                MM[MMnearAM.index[i]]=t
                AM[MMnearAM.index[i]]=z-t
                s= s+ 2*t
            else:
                MM[MMnearAM.index[i]]=z
                del AM[MMnearAM.index[i]]
                s= s +2*z
        i = i + 1

    # Having placed all the workers into the 5 slots we now compute the Total Commute Time for the workforce.
    # First we compute the total commute time for the workers in shift AM

    TotCommAM=[]
    for i in AM.keys():
        w=dJD.loc[i,"AtoM"]
        y=AM[i]
        m= y*w
        TotCommAM.append(m)


    SumAM=sum(TotCommAM)

    # We compute the total commte time for the workers in shift MP

    TotCommMM=[]

    for i in MM.keys():
        w=dJD.loc[i,"MtoM"]
        y=MM[i]
        m=y*w
        TotCommMM.append(m)


    SumMM=sum(TotCommMM)

    TotCommMN=[]
    for i in MN.keys():
        w=dJD.loc[i,"MtoN"]
        y=MN[i]
        m= y*w
        TotCommMN.append(m)


    SumMN=sum(TotCommMN)


    TotCommEA=[]
    for i in EA.keys():
        w=dJD.loc[i,"EtoA"]
        y=EA[i]
        m= y*w
        TotCommEA.append(m)


    SumEA=sum(TotCommEA)


    TotCommNN=[]
    for i in NN.keys():
        w=dJD.loc[i,"NtoN"]
        y=NN[i]
        m= y*w
        TotCommNN.append(m)


    SumNN=sum(TotCommNN)

    SumEAMMN_AMMNN = SumEA + SumAM + SumMM + SumMN + SumNN
    return SumEAMMN_AMMNN, 0, AM, 0, 0, MN, NN, EA,MM,0







def emmpn_mmpnn(WorkerDistrictID):
    
    Remainder = len(WorkerDistrictID) % 5
    TotalNumWorkers = len(WorkerDistrictID) - Remainder
    
    np.random.choice(WorkerDistrictID, len(WorkerDistrictID), replace=False)
    
    np.random.seed(10)
    neighborhood_List = np.random.choice(WorkerDistrictID, len(WorkerDistrictID), replace=False)
    
    
    Neighborhoodlist = neighborhood_List[0:TotalNumWorkers]
    

    
    
    NeighborhoodSeries = pd.Series(Neighborhoodlist)
    dJD = EMMPN_MMPNN.copy()
    dJD.insert(0,'NumWorkers', NeighborhoodSeries.value_counts())
    dJD=dJD.fillna(0)
    
    
    #We will first fill up the two shifts of EMMPN_MMPNN that have the lowest total commute time, EM and NN, with workers
    #in the districts that take the greatest values during these shifts, i.e. districts with the greatest values of DIFFMAX.
    #That is, we will count out starting from the bottom of EMMPN_MMPNN, 2/5*TotalNumWorkers who we will assign to EM and NN 
    #in the manner indicated below.


    j,b,d = g(2/5,dJD,TotalNumWorkers)
    
    #In counting 2/5*TotalNumWorkers workers starting from the bottom of EMMPN_MMPNN and working upwards, sometimes the number 
    #2/5*TotalNumWorkers may include some of the workers in the last district counted but not all of the workers in that district.
    #So that some of the workers in the last district considered are in the 2/5*TotalNumWorkers and therefore to be assigned
    #to EM or NN and the remainder of workers in that district will be assigned to the other three shifts.
    # 'b' denotes the number of workers in that last district that are counted in 2/5*TotalNumWorkers, and 'd' denotes the 
    # number of workers remaining in that last district that are not counted in 2/5*TotalNumWorkers. 'b+d' is the sum of the 
    # workers in that last district reached in counting 2/5*TotalNumWorkers.
    # The data frame EMMPN_MMPNN is broken up into two data frames: Low2 contains the 2/5*TotalNumWorkers workers who will be 
    # assigned shifts EM and NN; uppdJD wil contain the 3/5*TotalNumWorkers left which will include the d workers descrribed
    # above.
    

    



    ddJD= dJD[j:35]
    ddJD.copy()
    
    
    Low2= ddJD[['NumWorkers','District', 'EtoM', 'NtoN', 'DIFFMtoP']]
    Low2=Low2.copy()
    
    
    if b>0:
        Low2.iloc[0,0]= b
    else:
        Low2.iloc[0,0]=dJD.iloc[j,0]



    uppdJD= dJD[0:j+1]
    uppdJD=uppdJD.copy()
    

    uppdJD.iloc[j,0]= d

    NN={}
    EM={}

    for i in range(0,len(Low2)):
        a=Low2.iloc[i,1]
        b=Low2.iloc[i,0]
        if Low2.iloc[i,2]<Low2.iloc[i,3]:
            EM[a]=b

        else: 
            NN[a]=b


        i = i + 1
    
    DiffNtoNEtoM= Low2['NtoN']-Low2['EtoM']
    DiffNtoNEtoM[list(EM.keys())].sort_values()

    DiffEtoMNtoN= Low2['EtoM']-Low2['NtoN']
    DiffEtoMNtoN[list(NN.keys())].sort_values()

    EMnearNN=DiffEtoMNtoN[list(NN.keys())].sort_values()
    NNnearEM=DiffNtoNEtoM[list(EM.keys())].sort_values()
    

    r= 1/5*TotalNumWorkers-sum(EM.values())
    y= sum(NN.values())-1/5*TotalNumWorkers
    if r>0 and y>0:

        for i in range(0,len(EMnearNN)):
            x=EMnearNN.index[i]
            z=NN[x]
            w=y-z
            if w<0:

                NN[x]=z-y
                EM[x]=y
                break
            else:
                y=y-z
                EM[x]=z
                del NN[x]

    elif r<0 and y<0:
        t=-y
        for i in range(0,len(NNnearEM)):
            x=NNnearEM.index[i]
            z= EM[x]
            w= t-z
            if w<0:

                EM[x]=z-t
                NN[x]=t
                break
            else:
                t=t-z
                NN[x]=z
                del EM[x]

    
    #The possible 1, 2, 3 or 4 workers remaining upon division of TotalNumWorkers by 5, are all put into one of the two shifts 
    # with lowest total commute: NN. Although putting them into EM may yeild a lower total commute for this small sum of workers, 
    # we stick to NN for each of the 4 functions representing a different scheduling phrase because they each have an NN shift and
    # keeping this uniform for all 4 functions is useful as our objective is to determine which schedule function yields the 
    # lowest total commute time for the workforce. At the end, once a function is deemed the winner, an adjustment could be made
    # for these 1-4 possible workers if there is a shift EM. As these are at most 4 workers, this time difference is a tiny 
    # fraction of the total commute time of the workforce.
    
    neighborhood_List[TotalNumWorkers:]
    
    for i in neighborhood_List[TotalNumWorkers:]:
        if i in NN.keys():
            NN[i] = NN[i] + 1
        else:
            NN[i] = 1

   
    # Working now from the top of uppdJD which is sorted so as to be increasing in value of DIFFMtoP, we assign the workers 
    # in districts indicated in columns 0 and 1 the shift that has the highest values for total commute time: MP.
    #(In all but 4 out of the 35 districts, the total commute time is largest, bigger than that of PN and MM.) 
    # Next, as we move down uppdJD we fill workers into PN. PN has a higher total commute value than MM in all but 4 districts.
    # Continuing to move down uppdJD, we assign the remaining workers to MM.


    PN={}
    MP={}
    i=0
    l=0
    k=0
    count=0
    while count<1/5*TotalNumWorkers and i in range(0,j+1):
        a = uppdJD.iloc[i,0]
        b = uppdJD.iloc[i,1]
        count= a + count
        if count>1/5*TotalNumWorkers:
            l= count-1/5*TotalNumWorkers
            m=uppdJD.iloc[i,0]-l
            MP[b]=m
            PN[b]=l
        else:
            PN[b]=a

        i = i + 1


    MM={}
    k=i
    count=0
    while count in range(0,int(1/5*TotalNumWorkers)-int(l)) and k in range(i,j+1):
        a = uppdJD.iloc[k,0]
        b = uppdJD.iloc[k,1]
        count = a + count
        if count> 1/5*TotalNumWorkers-l:
            r= count-(1/5*TotalNumWorkers-l)
            s= uppdJD.iloc[k,0]-r
            PN[b]=s
            MM[b]=r
        else:
            PN[b]=a

        k = k + 1


    h=k
    count=2/5*TotalNumWorkers
    u=int(2/5*TotalNumWorkers)
    v=int(3/5*TotalNumWorkers)
    while count in range(u,v) and h in range(k,j+1):
        a = uppdJD.iloc[h,0]
        b = uppdJD.iloc[h,1]
        MM[b]=a
        count = a + count 

        h = h + 1


    TotCommEM=[]
    for i in EM.keys():
        w=dJD.loc[i,"EtoM"]
        y=EM[i]
        m= y*w
        TotCommEM.append(m)

    
    SumEM=sum(TotCommEM)
    

    TotCommPN=[]
    for i in PN.keys():
        w=dJD.loc[i,"PtoN"]
        y=PN[i]
        m=y*w
        TotCommPN.append(m)
    
    SumPN=sum(TotCommPN)
    

    TotCommMM=[]
    for i in MM.keys():
        w=dJD.loc[i,"MtoM"]
        y=MM[i]
        m=y*w
        TotCommMM.append(m)
    
    SumMM=sum(TotCommMM)
    
    TotCommMP=[]
    for i in MP.keys():
        w=dJD.loc[i,"MtoP"]
        y=MP[i]
        m=y*w
        TotCommMP.append(m)
    
    SumMP=sum(TotCommMP)
    
    TotCommNN=[]
    for i in NN.keys():
        w=dJD.loc[i,"NtoN"]
        y=NN[i]
        m=y*w
        TotCommNN.append(m)
    
    SumNN=sum(TotCommNN)
    
    SumEMMPN_MMPNN = SumEM + SumPN + SumMM + SumMP + SumNN
    return SumEMMPN_MMPNN, EM,0,MP,0,0,NN,0,MM,PN


```
