```python
def g(dJD,TotalNumWorkers):
    i=0
    count=0
    b=0
    d=0
    while count<0.7*TotalNumWorkers:
        j= len(dJD)-1-i
        newcount= dJD.iloc[j,0]
        count= newcount + count
        if count>0.7*TotalNumWorkers:
                d= count-0.7*TotalNumWorkers
                b=dJD.iloc[j,0]-d
        i = i + 1
        
    return j,b,d

def emm_mmn(WorkerDistrictID):
    
    
    Remainder = len(WorkerDistrictID) % 5
    TotalNumWorkers = len(WorkerDistrictID) - Remainder
    
    np.random.seed(10)
    neighborhood_List = np.random.choice(WorkerDistrictID, len(WorkerDistrictID), replace=False)
    
    
    Neighborhoodlist = neighborhood_List[0:TotalNumWorkers]
   
    
    

    NeighborhoodSeries = pd.Series(Neighborhoodlist)
    dJD=EMM_MMN.copy()
    dJD.insert(0,'NumWorkers', NeighborhoodSeries.value_counts())
    dJD=dJD.fillna(0)
    
    
    #We will first fill up the two shifts of EMM_MMN that have the shortest commutes, EM, NN, with workers in the districts
    #that take the greatest values during these shifts, i.e. districts with the greates values of DIFFMAX.That is, we will count
    #out starting from the bottom of EMM_MMN, 0.7*TotalNumWorkers who we will assign to EM and NN in the manner indicated 
    #in the code that follows.
    
    j,b,d = g(dJD,TotalNumWorkers)
    
    
    
    #In counting 0.7*TotalNumWorkers workers starting from the bottom of EMM_MMN and working upwards, sometimes the number 
    #0.7*TotalNumWorkers may include some of the workers in the last district counted but not all of the workers in that district.
    #So that some of the workers in the last district considered are in the 0.7*TotalNumWorkers and therefore to be assigned
    #to EM or MN, and the remainder of workers in that district will be assigned to the other shift.
    # 'b' denotes the number of workers in that last district that are counted in 0.7*TotalNumWorkers, and 'd' denotes the 
    # number of workers remaining in that last district that are not counted in 0.7*TotalNumWorkers. 'b+d' is the sum of the 
    # workers in that last district reached in counting 0.7*TotalNumWorkers.
    # The data frame EMM_MMN is broken up into two data frames: Low2 contains the 0.7*TotalNumWorkers workers who will be 
    # assigned shifts EM and MN; uppdJD wil contain the 0.3*TotalNumWorkers left which will include the d workers descrribed
    # above.
    
    Low2= dJD[j:35].copy()
    
    
    if b>0:
        Low2.iloc[0,0]=b
    
    uppdJD= dJD[0:j+1]
    uppdJD= uppdJD.copy()
   
    
    if d>0:
        uppdJD.iloc[j,0]= d
    else:
        uppdJD.iloc[j,0]=0
        
    EM={}
    MN={}
    for i in range(0,len(Low2)):
        a=Low2.iloc[i,1]
        b=Low2.iloc[i,0]
        if Low2.iloc[i,2]<Low2.iloc[i,5]:
            EM[a]=b

        else:
            MN[a]=b
        i = i + 1
    
   
    
    DiffEMandMN= Low2['MtoN']-Low2['EtoM']
    DiffEMandMN[list(EM.keys())]
    DiffEMandMN[list(EM.keys())].sort_values()
    
    EMnearMN=DiffEMandMN[list(EM.keys())].sort_values()

  
    y=sum(EM.values())-0.35*TotalNumWorkers
 
    for i in range(0,len(EMnearMN)):
        x=EMnearMN.index[i]
        z=EM[x]
        w=y-z
        if w<0:

            EM[x]=z-y
            MN[x]=y
            break
        else:
            y=y-z
            MN[x]=z
            del EM[x]
   
    
   
    
    
    for i in neighborhood_List[TotalNumWorkers:]:
        if i in EM.keys():
            EM[i] = EM[i] + 1
        else:
            EM[i] = 1

    
    MM={}

    for i in range(0,j+1):
        a=uppdJD.iloc[i,1]
        b=uppdJD.iloc[i,0]

        MM[a]=b


   
       
    TotCommMM=[]

    for i in MM.keys():
        w=dJD.loc[i,"MtoM"]
        y=MM[i]
        m=y*w
        TotCommMM.append(m)

 
    SumMM=sum(TotCommMM)
    
    
    TotCommMN=[]
    for i in MN.keys():
        w=dJD.loc[i,"MtoN"]
        y=MN[i]
        m= y*w
        TotCommMN.append(m)

 
    SumMN=sum(TotCommMN)
  
    
    TotCommEM=[]
    for i in EM.keys():
        w=dJD.loc[i,"EtoM"]
        y=EM[i]
        m= y*w
        TotCommEM.append(m)

  
    SumEM=sum(TotCommEM)
   
    
    SumEMM_MMN = SumEM + SumMM + SumMN 
    return SumEMM_MMN, EM, MN, MM, 0



```

```python
def eam_mmn(WorkerDistrictID):
    
    
    Remainder = len(WorkerDistrictID) % 5
    TotalNumWorkers = len(WorkerDistrictID) - Remainder
    
    np.random.seed(10)
    neighborhood_List = np.random.choice(WorkerDistrictID, len(WorkerDistrictID), replace=False)
    
    
    Neighborhoodlist = neighborhood_List[0:TotalNumWorkers]
 
    
    

    NeighborhoodSeries = pd.Series(Neighborhoodlist)
    dJD=EMM_MMN.copy()
    dJD.insert(0,'NumWorkers', NeighborhoodSeries.value_counts())
    dJD=dJD.fillna(0)
    
    
    #We will first fill up the two shifts of EAM_MMN having the shortest commutes, EM and MN, with workers in the districts
    #that take the greatest values during these shifts, i.e. districts with the greates values of DIFFMAX. That is, we will 
    #count out starting from the bottom of EAM_MMN, 0.7*TotalNumWorkers who we will assign to EM and MN, in the manner indicated 
    #ahead.
    
    j,b,d = g(dJD,TotalNumWorkers)
    
    
    
    #In counting 0.7*TotalNumWorkers workers starting from the bottom of EAM_MMN and working upwards, sometimes the number 
    #0.7*TotalNumWorkers may include some of the workers in the last district counted but not all of the workers in that district.
    #So that some of the workers in the last district considered are in the 0.7*TotalNumWorkers and therefore to be assigned
    #to EM or MN, and the remainder of workers in that district will be assigned to the other shift.
    # 'b' denotes the number of workers in that last district that are counted in 0.7*TotalNumWorkers, and 'd' denotes the 
    # number of workers remaining in that last district that are not counted in 0.7*TotalNumWorkers. 'b+d' is the sum of the 
    # workers in that last district reached in counting 0.7*TotalNumWorkers.
    # The data frame EAM_MMN is broken up into two data frames: Low2 contains the 0.7*TotalNumWorkers workers who will be 
    # assigned shifts EM and MN; uppdJD wil contain the 0.3*TotalNumWorkers left which will include the d workers descrribed
    # above. The workers in uppdJD wil be assigned to the remaining shift, AM.
    
    Low2= dJD[j:35].copy()
    
    
    if b>0:
        Low2.iloc[0,0]=b
    
    uppdJD= dJD[0:j+1]
    uppdJD= uppdJD.copy()
   
    
    if d>0:
        uppdJD.iloc[j,0]= d
    else:
        uppdJD.iloc[j,0]=0
        
    EM={}
    MN={}
    for i in range(0,len(Low2)):
        a=Low2.iloc[i,1]
        b=Low2.iloc[i,0]
        if Low2.iloc[i,2]<Low2.iloc[i,5]:
            EM[a]=b

        else:
            MN[a]=b
        i = i + 1
  
    
   
    
    DiffEMandMN= Low2['MtoN']-Low2['EtoM']
    DiffEMandMN[list(EM.keys())]
    DiffEMandMN[list(EM.keys())].sort_values()
    
    EMnearMN=DiffEMandMN[list(EM.keys())].sort_values()

   
    y=sum(EM.values())-0.35*TotalNumWorkers
   
    for i in range(0,len(EMnearMN)):
        x=EMnearMN.index[i]
        z=EM[x]
        w=y-z
        if w<0:

            EM[x]=z-y
            MN[x]=y
            break
        else:
            y=y-z
            MN[x]=z
            del EM[x]
    
    
    
    
    for i in neighborhood_List[TotalNumWorkers:]:
        if i in EM.keys():
            EM[i] = EM[i] + 1
        else:
            EM[i] = 1

  
    
    AM={}

    for i in range(0,j+1):
        a=uppdJD.iloc[i,1]
        b=uppdJD.iloc[i,0]

        AM[a]=b

    
       
    TotCommAM=[]

    for i in AM.keys():
        w=dJD.loc[i,"AtoM"]
        y=AM[i]
        m=y*w
        TotCommAM.append(m)

  
    SumAM=sum(TotCommAM)
  
    
    TotCommMN=[]
    for i in MN.keys():
        w=dJD.loc[i,"MtoN"]
        y=MN[i]
        m= y*w
        TotCommMN.append(m)

  
    SumMN=sum(TotCommMN)
  
    
    TotCommEM=[]
    for i in EM.keys():
        w=dJD.loc[i,"EtoM"]
        y=EM[i]
        m= y*w
        TotCommEM.append(m)

  
    SumEM=sum(TotCommEM)
 
    
    SumEAM_MMN = SumEM + SumAM + SumMN 
    return SumEAM_MMN, EM, MN, 0, AM



```
